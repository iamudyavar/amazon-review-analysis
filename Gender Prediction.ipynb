```python
import joblib

# Load the pre-trained model
loaded_model = joblib.load('gender_predictor.pkl')

# Example name for prediction
new_name = "queenie"
new_name = [new_name.lower()]
print(new_name)

# Use the loaded model for prediction
predicted_gender = loaded_model.predict(new_name)

# Display the predicted gender (0 for male, 1 for female)
if predicted_gender[0] == 0:
    print("Predicted Gender: Male")
else:
    print("Predicted Gender: Female")
```

```python
import pandas as pd
import joblib

df = pd.read_csv("Filtered Reviews.csv")
df

loaded_model = joblib.load('gender_predictor.pkl')
```

```python
# Predict gender
def predict_gender(name):
    predicted_gender = loaded_model.predict([name.lower()])
    if predicted_gender[0] == 0:
        return("Male")
    else:
        return("Female")
    
# Predict gender of each review
df['Gender'] = df.apply(lambda column: predict_gender(column['ProfileName']), axis=1)

male_count = len(df[df['Gender'] == 'Male'])
print(male_count)
df
```

```python
df.to_csv('Gender.csv', index=False)
```
